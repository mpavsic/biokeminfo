### Funkcija proteina
Sodeluje pri biosintezi aminokislin (sinteza L-Thr iz L-Asp), protein je vključen v pretvorbo aspartata (vključen v sintezo esencialnih aminokislinskoh ostankov). 
Je ATP-vezavni protein; vezavno mesto za ATP je na mestih 91-101. ![ATP-vezavno_mesto](s46-vezavno.png) Kot pove že ime, ima homoserin-kinazno aktivnost (je encim skupine transferaz); katalizira od ATP odvisno fosforilacjo L-homoserina v L-homoserin fosfat. Katalitska aktivnost je 500-krat višja z L-homoserinom kot D-homoserinom. Fosforilira tudi hidroksilno skupino gama-ogljika L-homoserinskih analogov, pri katerih je funkcionalna skupina na mestu alfa karboksilna, estrska ali hidroksimetilna. 
Protein je katalitsko aktiven (kinazna aktivnost; prenos fostatne skupine (donorska molekula je v tem primeru ATP) na substrat (akceptorji so alkoholne skupine)), katalizira reakcijo ATP + L-homoserin = ADP + H+ + O-fosfo-L-homoserin.![Omenjena_reakcija](s46-reakcija_kat.png)

### Sorodni proteini
Spada v družino GHMP kinaz, ki nasploh obsega od ATP-odvisne kinaze; vključuje homoserin kinaze, galaktokinaze, mevalonat kinaze - te predvsem sintetizirajo izoprene in aminokisline, sodelujejo pa tudi v določenih drugih metabolnih procesih. Homoserin kinaza, kodirana z genom thrB je poleg pri E. coli prisotna tudi pri mnogih drugih mikroorganizmih (npr. *Staphylococcus aureus, Mycobacterium sp. Bacillus subtilis, Clostridium tetani ...*), med katerimi je veliko patogenih (morda se bo lahko začeo razmišljati v smeri potencialne tarče). Te podatke in proteine omenjene družine lahko poiščemo na Uniprotu.
Če z orodjem EMBOSS Dotmacher primerjamo zaporedje homoserin kinaze iz organizmov E.Coli in iz *Methanocaldococcus jannaschii* (izvorni organizem deponirane strukture v PDB). Opazimo veliko podobnost (izrazit vzorec diagonale z nekaj prazninami - ostanki prisotni le pri enem izmed zaporedij). ![Primerjava_homoserin_kinaz](s46-emboss_dot.png)
V družino GHMP kinaz spada tudi galaktokinaza, ki je prav tako prisotna pri E.coli sevu 12, protein najdemo na Uniprotu. Z zanimanjem preverimo tudi podobnost tega proteina našemu, tako obe zaporedji vnesemo v Dotmatcher, rezultat pa pokaže, da si proteina nista podobna. ![Primerjava_nepodobnih_dveh](s46-emboss_nepodobna.png)  
Če izvedemo iskanje s PSI-blastom, opazimo podobnosti s hipotetičnimi protieni GOS (isti metagenom) in proteini morskih sedimentov, a zadetki niso zelo relevantni. Zato spremenimo parametre in iščemo le po podatkovni bazi dobro anotiranih proteinov (Swissprot/Uniprot), namesto po metagenomski zbirki. ![Sprememba_parametrov](s46-psi_parametri.png) Orodje zaženemo npr. 7-krat (v mojem primeru), novi rezultati se še dalje pojavljajo. Če poravnana zaporedja prenesemo, nam to omogoča nadaljnje analize.
V Uniprotu izberemo 10 zaporedij za homoserin kinazo pri različnih mikroorganizmih, katera uporabimo pri delu z Orodjem Clustal Omega.![Izsek podatkov iz Clustal Omega](s46-clustal.png) Orodje nam poda podatke za izdelavo filogenetskega drevesa z orodjem phylo.io, s čimer si lahko ogledamo potek evolucije preučevanega proteina (od skupnih prednikov do današnjih). ![Filogenetsko_drevo](s46-phylo.png)
V Uniprotu razberemo, da je homoserin kinaza prisotna tudi pri evkationtih; rižu (*Oryza sativa subsp. japonica*), kjer opravlja isto funkcijo, a struktura ni eksperimentalno določena (slabša anotacija, podani trije rezultati, Alphafold struktura je predvidena z manjšo zanesljivostjo). Protein ima določeni sicer enaki domeni kot izhodiščna homoserin kinaza. Če direktno v Uniprotu primerjamo po eno zaporedje z našim vidimo, da prihaja do ohranjenosti določenih delov, pri čem ne moremo z zanesljivostjo sklepati, če se ohranja aktivno mesto, saj v primeru riža, to ni z gotovostjo določeno. Priložena je slika izseka poravnave, kjer se sicer nahaja aktivno mesto proteina pri E.coli. ![Uniprot_poravnava](s46-poravnava_riz.png)

### Ohranjenost in mutacije
Če v Uniprotu pregledamo homoserin kinaze večih organizmov vidimo, da se med organizmi zagotovo ohranja ATP-vezavno mesto, saj se s tem ohranja funkcija proteina (vezava ATP, ki omogoča kinazno aktivnost). 
Prisotnost mutacij vpliva na katalitsko učinkovitost encima (mutacija na mestu 235: R->L: zmanjšana učinkovitost, R->H povečana) in kinazno aktivnost (mesta 139 in 203). Tej podatki so dostopni na Uniprotu. ![Prikaz_mutacij](s46-mutacije.png)
Ohranjene domene med organizmi identificiramo s pomočjo orodja COBALT. Z orodjem analiziramo zaporedja, do katerih pridemo s PSI-blastom (naložimo v nekompresirani obliki). Izriše se nam grafični prikaz, kjer je opazno, da prihaja do signifikantne ohranjenosti med uporabljenimi zaporedji, seveda pa tudi do razlik (glede na vzorec grafa, obarvanost glede na ohranjenost; rdeči predeli označujejo signifikantno ohranjenost regij, modri pa slabšo - slika prikazuje odsek začetnih nekaj poravnav). ![Poravnave-cobalt](s46-cobalt.png)
